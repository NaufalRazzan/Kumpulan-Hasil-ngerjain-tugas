2). handling missing data (Replacing missing data with the mean value)

```python
# melihat total NaN
df.isnull().sum()
```

```python
# drop kolom name yang ada NaN
df1 = df.dropna(subset = ['name'])
df1
```

```python
# drop kolom last review
df1.drop('last_review', axis = 1, inplace = True)
df1
```

```python
# ganti NaN di kolom review per month dengan mean
mean_value = df1['reviews_per_month'].mean()
df1['reviews_per_month'].fillna(value = mean_value, inplace = True)
df1
```

3). Encoding Categorical data for neighbourhood_group variable and room_type variable

```python
from sklearn.preprocessing import LabelEncoder
# x -> neighbourhood_group, y -> room_type
x = y = LabelEncoder()
df1.iloc[:, 4] = x.fit_transform(df1.iloc[:, 4])
df1.iloc[:, 8] = y.fit_transform(df1.iloc[:, 8])

df1
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# ct1 -> x, ct2 -> y
ct1 = ColumnTransformer(transformers = [('en1', OneHotEncoder(), [4])], remainder = 'passthrough')
ct2 = ColumnTransformer(transformers = [('en2', OneHotEncoder(), [8])], remainder = 'passthrough')

df2 = ct1.fit_transform(df1) 
df2 = ct2.fit_transform(df1)

df2 = pd.DataFrame(df2)
df2
```
