```python
import pandas as pd

ls = pd.read_csv('listings1.csv')
ls.head()
```

1). covariance dan correlation antara 2 kolom

```python
# saya akan mengambil 4 kolom acak sebagai sampel kecil (2 kolom categorical dan 2 kolom numeric)
ls1 = ls.iloc[:, [3, 8, 9, 10]]
ls1
```

```python
# encode categorical data
from sklearn.preprocessing import LabelEncoder
# x -> Host_name, y -> room_type
x = y = LabelEncoder()
ls1.iloc[:, 0] = x.fit_transform(ls1.iloc[:, 0])
ls1.iloc[:, 1] = y.fit_transform(ls1.iloc[:, 1])

ls1
```

```python
# covariance & correlation categorical vs categorical
print('Covariance:')
print(ls1.iloc[:, [0, 1]].cov())
print('\nCorrelation:')
print(ls1.iloc[:, [0, 1]].corr())
```

Antara dua kolom diatas, terdapat covariance yang positif. Sementara correlation juga positif. Jika di lihat, kedua nilai covariance dan correlation cukup kecil. Artinya, kedua kolom memiliki hubungan yang lemah.

```python
# covariance & correlation categorical vs numeric
print('Covariance:')
print(ls1.iloc[:, [0, 3]].cov())
print('\nCorrelation:')
print(ls1.iloc[:, [0, 3]].corr())
```

Antara dua kolom diatas, terdapat covariance dan correlation yang negatif. Meskipun covariance sangat besar (mengindikasikan hubungan yang sangat kuat), tetapi correlation cukup kecil. 

```python
# covariance & correlation numeric vs numeric
print('Covariance:')
print(ls1.iloc[:, [2, 3]].cov())
print('\nCorrelation:')
print(ls1.iloc[:, [2, 3]].corr())
```

Antara dua kolom diatas, terdapat covariance dan correlation yang negatif.

3). Feature selection & Extraction

```python
# menggunakan dataframe ls1
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

array = ls1.values
x = array[:, 0:3]
y = array[:, 3]
# feature selection
test = SelectKBest(score_func = f_classif, k = 2)
fit = test.fit(x, y)
# score summary
print(fit.scores_)
```
