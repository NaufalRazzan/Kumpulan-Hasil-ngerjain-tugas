**clustering menggunakan kmodes**

```python
df
```

```python
from kmodes.kmodes import KModes
import numpy as np

# elbow method untuk mencari nilai k optimal

cu = []

df_ml = np.array(df_ml)

for i in range(1, 11):
    kmode = KModes(n_clusters=i, init="random", n_init=5, verbose=1)
    kmode.fit_predict(df_ml)
    cu.append(kmode.cost_)

plt.figure(figsize=(5, 3))
plt.plot(range(1, 11), cu, 'bx-')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('The Elbow Method')
plt.show()
```

```python
df_ml
```

```python
# visualisasi kmodes dengan 4 cluster

km_visual = KModes(n_clusters=4, init="random", n_init=5, verbose=1)
km_output = km_visual.fit_predict(df_ml)

df['cluster'] = km_output
```

```python
# visualisasi

for col in df.columns:
    plt.subplots(figsize=(5,3))
    sns.countplot(x='cluster', hue=col, data=df)
    plt.show()
```
