```python
import pandas as pd

df = pd.read_csv('listings2.csv')
df
```

1). Extracting discrete and continuous random variables

```python
x = df[['id', 'name', 'host_id', 'host_name', 'neighbourhood_group', 'neighbourhood', 'room_type', 'price', 'minimum_nights', 'number_of_reviews', 'last_review', 'calculated_host_listings_count', 'availability_365']]
y = df[['latitude', 'longitude', 'reviews_per_month']]
```

```python
x
```

```python
y
```

2). Calculated PMF (probability mass function) for discrete random variable

```python
# variable = room_type

pmf_list = pd.concat([x.room_type.value_counts(normalize = True, ascending = True), x.room_type.value_counts()], axis = 1, keys= ('pmf', 'count'))

pmf_list
```

3). Plot bar chart

```python
pmf = x.room_type.value_counts(normalize = True, ascending = True)
pmf.plot.bar()
```

4). Calculated PDF (probability density function) for continuous random variables

```python
# variable = longitude

pdf = y.longitude.value_counts(normalize = True, ascending = True)
pdf
```

```python
pdf_list = pd.concat([pdf, y.longitude.value_counts(ascending = True)], axis = 1, keys = ('pdf', 'count'))
pdf_list
```

5). Plot Histogram

```python
pdf_list.hist()
```
